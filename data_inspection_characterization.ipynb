# Inspect and Characterize Data
Here we're just looking for any obvious problems with the data and beginning to characterize it.  
Some basic info includes the number of data points per trial, frequency content, intensity histograms, mean/median/maximum values, etc.

```python
# clear vars
%reset -f

# show plots in the notebook
%matplotlib inline

# reload modules automatically
%load_ext autoreload
%autoreload 2

import os              # directory walking 
import pandas as pd    # data manipulation

from athospy.plt_config import mpl_plt
mpl, plt = mpl_plt()

DATA_DIR = '../data/cleaned/anonymized/'
RAW_FIGS_DIR = '../data/results/figs/inspect_raw'
try:
    os.mkdir(RAW_FIGS_DIR)
except:
    print('already exists')
```

Define and use a plotting function to show left, right, and accelerometer data

```python
def df_line(df, title):
    df_emgR = df.loc[:, ['RGM','RBF','RVM','RVL']]
    df_emgL = df.loc[:, ['LGM','LBF','LVM','LVL']]
    df_acc = df.loc[:,'ACCX':'ACCZ']
    
    fig, ax = plt.subplots(3, 1, sharex=True)
    df_emgR.plot(ax=ax[0]).legend(loc='center left', bbox_to_anchor=(1, 0.5))
    df_emgL.plot(ax=ax[1]).legend(loc='center left', bbox_to_anchor=(1, 0.5))
    df_acc.plot(ax=ax[2]).legend(loc='center left', bbox_to_anchor=(1, 0.5))

    ax[0].set_ylabel('Right EMG')
    ax[1].set_ylabel('Left EMG')
    ax[2].set_ylabel('Accel.')
    
    plt.suptitle(title)
    
    return fig, ax
```

```python
# TODO: alpha for transparency, refactor
def df_hist(df, title):
    df_emgR = df.loc[:, ['RGM','RBF','RVM','RVL']]
    df_emgL = df.loc[:, ['LGM','LBF','LVM','LVL']]
    
    prms = {'kind': 'hist',
            'alpha': 0.25,
            'ylim': [0,300],
            'logy': True
            }
    
    fig, ax = plt.subplots(2, 1, sharex=True)
    df_emgR.plot(ax=ax[0], **prms).legend(
        loc='center left', bbox_to_anchor=(1, 0.5))
    df_emgL.plot(ax=ax[1], **prms).legend(
        loc='center left', bbox_to_anchor=(1, 0.5))

    ax[0].set_ylabel('Right EMG')
    ax[1].set_ylabel('Left EMG')
    
    plt.suptitle(title)
    
    return fig, ax
```

```python
def df_load(fname):
    fname, ext = os.path.splitext(fname)
    if ext:
        assert ext == '.csv', "extension must be .csv, not %s" % ext
    fn = os.path.join(DATA_DIR, fname + '.csv')
    
    df = pd.read_csv(fn, header=0)
    return df
```

## Inspect the entire dataset
For each .csv file loaded with `df_load`, not `df_load_emg`
* save a plot of the raw data
* calculate basic statistics (number of rows, max, mean, median

First, load and plot the raw data

```python
isSaving = False
empty_files = []
for fn in os.listdir(DATA_DIR):
    fname, _ = os.path.splitext(fn)
    df = df_load(fname)
    sz = df.size
    if sz:
        if isSaving:
            fig, ax = df_line(df, fname)
            fig.savefig(os.path.join(RAW_FIGS_DIR, fname + '.pdf'))
            plt.close() # prevents figure from being plotted
    else:
        empty_files.append(fname)
```
