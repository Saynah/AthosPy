Define and use a plotting function to show left, right, and accelerometer data

```python
def df_line(df, title):
    df_emgR = df.loc[:, ['RGM','RBF','RVM','RVL']]
    df_emgL = df.loc[:, ['LGM','LBF','LVM','LVL']]
    df_acc = df.loc[:,'ACCX':'ACCZ']
    
    fig, ax = plt.subplots(3, 1, sharex=True)
    df_emgR.plot(ax=ax[0]).legend(loc='center left', bbox_to_anchor=(1, 0.5))
    df_emgL.plot(ax=ax[1]).legend(loc='center left', bbox_to_anchor=(1, 0.5))
    df_acc.plot(ax=ax[2]).legend(loc='center left', bbox_to_anchor=(1, 0.5))

    ax[0].set_ylabel('Right EMG')
    ax[1].set_ylabel('Left EMG')
    ax[2].set_ylabel('Accel.')
    
    plt.suptitle(title)
    
    return fig, ax
```

```python
# TODO: alpha for transparency, refactor
def df_hist(df, title):
    df_emgR = df.loc[:, ['RGM','RBF','RVM','RVL']]
    df_emgL = df.loc[:, ['LGM','LBF','LVM','LVL']]
    
    prms = {'kind': 'hist',
            'alpha': 0.25,
            'ylim': [0,300],
            'logy': True
            }
    
    fig, ax = plt.subplots(2, 1, sharex=True)
    df_emgR.plot(ax=ax[0], **prms).legend(
        loc='center left', bbox_to_anchor=(1, 0.5))
    df_emgL.plot(ax=ax[1], **prms).legend(
        loc='center left', bbox_to_anchor=(1, 0.5))

    ax[0].set_ylabel('Right EMG')
    ax[1].set_ylabel('Left EMG')
    
    plt.suptitle(title)
    
    return fig, ax
```

```python
def df_load(fname):
    fname, ext = os.path.splitext(fname)
    if ext:
        assert ext == '.csv', "extension must be .csv, not %s" % ext
    fn = os.path.join(DATA_DIR, fname + '.csv')
    
    df = pd.read_csv(fn, header=0)
    return df
```

```python
def loadAndLine(fname):
    df = df_load(fname)
    fig, ax = df_line(df, fname)
    return df
```

## Summary statistics for each file
Now, pull some summary statistics just for the EMG data to help identify bad data.
* number of rows - to find files that are too short or long enough to sample multiple times
* max value across all channels (below 30,000) - magnitude of the signal
* median value across all channels - find files with signals that are too low
* number of spikes to 65535 - files where sensor was freaking out
* max fraction of zero values across channels - possibly dead sensors?
* max fraction of consecutive non-zero values - files with temporal binning issues
* accel length relative to EMG length - figure out if ACCEL sampling rate can be recovered

```python
def csv_summary(fname):
    df = df_load_emg(fname) # load only EMG data
    
    len_df = len(df)
    
    d_summary = {
        "len_df" : len_df,
        "median_df" : df.unstack().median(),
        "max_df" : df[df<15000].unstack().max(),
        "n_spikes" : (df>65000).sum().sum(),
        "maxf_zeros" : max((df==0).sum().divide(len_df) * 100),
        "maxf_repeat" : max(((df!=0) & (df.diff()==0)).sum().divide(len_df) * 100)
    }
    return d_summary
```

Get summary statistics for each data file and create a table

```python
fnames, d_summ = [], []
for fn in os.listdir(DATA_DIR):
    fnames.append(fn)
    d_summ.append(csv_summary(fn))
    
df_summ = pd.DataFrame(d_summ, index=fnames)
```
