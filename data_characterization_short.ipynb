## Exclude .csv files with weird stats

So, picking only the top 5 exercises

Exclusion criteria:
* files < 500 in length (will sample 240 left and right -> 6 seconds)
* files with max frac of repeats > 60%
* files with max fraz zeros > 30%
* files with median values > 100[a.u.] (lots of noise)

```python
def exclude(df, is_plot=False):
    
    # exclude by exercise 1st
    df = df[(df.Exercise == 'LegPress') |
            (df.Exercise == 'Stairs') |
            (df.Exercise == 'ProneHamstringCurl') |
            (df.Exercise == 'ButtBlaster') |
            (df.Exercise == 'GluteBridge') ]
    
    len_old = len(df)
    
    d_summ = []
    for fname in df.index:
        fpath = os.path.join(DATA_DIR, fname)
        d_summ.append( fop.csv_summary(fpath) )

    df_summ = pd.DataFrame(d_summ, index=df.index)
    
    # plot
    if is_plot:
        y = np.random.randn(len(df_summ))
        fig, ax = plt.subplots(df_summ.shape[1])
        for i, axi in enumerate(ax):
            axi.scatter(df_summ.iloc[:,i], y)
            axi.set_title(df_summ.columns[i])

    short = df_summ.Length < 500 #500
    repeats = df_summ.MaxFrac_repeat > 60
    zeros = df_summ.MaxFrac_zero > 30
    noisy = df_summ.Median  > 100

    df = df[~short & ~repeats & ~zeros & ~noisy]
    print 'excluded %d files of %d' % (len_old - len(df), len_old)
    
    return df
```

```python
df_train = exclude(df_train, True)
df_test = exclude(df_test)
df47 = exclude(df47)
```

## Select 10 sec chunks from the middle of each file

```python
def build_df_dict(df, n_sec):
    n_samp = int(41.7 * n_sec)

    df_dict = {}
    for fname in df.index:
        df = fop.load_emg_df(DATA_DIR + fname)

        i_mid = df.shape[0] // 2
        i_start = i_mid - n_samp // 2
        i_end = i_mid + n_samp // 2

        df_dict[fname] = df.iloc[i_start:i_end]
    
    return df_dict
```

```python
df_dict_train = build_df_dict(df_train, 10)
df_dict_test = build_df_dict(df_test, 10)
```

```python
df_dict_47 = build_df_dict(df47, 10)
```

 ## PCA to reduce dimensions

```python
mpl.rc('figure', figsize=[10,10])
```

```python
def doPCA(f_df, df, title):

    pca = PCA(n_components=2)
    x_df = pca.fit(f_df).transform(f_df)
    x_df = pd.DataFrame(x_df,
                        index=f_df.index,
                        columns=['PC_1','PC_2'])

    print 'Percent variance explained:', pca.explained_variance_ratio_ * 100

    ix = np.argsort(pca.components_, axis=1)

    top = range(3)
    heavy1 = f_df.columns.values[ix[0,top]].tolist()
    heavy2 = f_df.columns.values[ix[1,top]].tolist()

    groups = x_df.join(df.Exercise).groupby('Exercise')

    fig, ax = plt.subplots()
    ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
    for name, group in groups:
        ax.plot(group.PC_1, group.PC_2,
                marker='.', linestyle='', ms=12, label=name)
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    ax.set_xlabel('PC_1 ~ ' + str(heavy1))
    ax.set_ylabel('PC_2 ~ '+ str(heavy2))
    ax.set_title(title)
    plt.show()
```

```python
doPCA(f_df_test, df_test, 'test')
```

## LDA

```python
lda = LDA(n_components=2)
x2_df = lda.fit(f_df_train, df_train.Exercise).transform(f_df_train)

x2_df = pd.DataFrame(x2_df,
                    index=f_df_train.index,
                    columns=['LDA_1','LDA_2'])

groups = x2_df.join(df_train.Exercise).groupby('Exercise')

fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
    ax.plot(group.LDA_1, group.LDA_2,
            marker='.', linestyle='', ms=12, label=name);
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5));
```

```python
def report(predicted, df, classes):
    labels = df.Exercise

    print '\npercent correct:', sum(predicted==labels) / len(predicted) * 100
    print metrics.classification_report(labels, predicted)
    
    counts = df.groupby('Exercise').count().values[:,0]
    
    mat = metrics.confusion_matrix(labels, predicted) / counts
    
    plot_confusion(mat, classes)
```

```python
def plot_confusion(mat, classes):
    
    mpl.rc('figure', figsize=[10, 10])

    fig, ax = plt.subplots()
    plt.pcolor(mat, cmap=plt.cm.Blues)

    # put the major ticks at the middle of each cell
    ax.set_xticks(np.arange(mat.shape[0])+0.5, minor=False)
    ax.set_yticks(np.arange(mat.shape[1])+0.5, minor=False)

    # want a more natural, table-like display
    ax.invert_yaxis()
    ax.xaxis.tick_top()

    # ax.set_xticklabels(row_labels, minor=False)
    # ax.set_yticklabels(column_labels, minor=False)

    ax.set_xticklabels(classes);
    ax.set_yticklabels(classes);
    
#??     ax.xaxis.label.set_fontsize(14)
```
