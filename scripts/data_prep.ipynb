# Data Preparation and Labeling
The goal here is to label the dataset based on person and exercise name, create a .json file that represents the labeled datasets (as a record of what was done), and also write out an anonymized dataset with the proper labels to file for downstream processing.

```python
import re # regular expressions
import os # directory walking
import fnmatch # extension matching
import difflib # fuzzy string match
import json # for dumping out dictionary
import matplotlib

DATA_DIR = '../../data/Research-Sessions/'
# EXER_LIST = '../config/exercise-list_long.txt'
EXER_LIST = '../config/exercise-list_short.txt'   # consolidated list
CLEAN_DIR = '../../data/cleaned/'

%matplotlib inline
```

Here are the functions for string parsing and matching. There might be a better alternative to `findall()`.
* person names match the pattern: `<first>_(<middle>:optional)_<last>_Calib ... .csv`
* exercise names match the pattern: `(<name>:ignore)_<exercise name> ... .csv`, where `"..."` is less than 7 characters.

```python
def extract_name(sub_dir):
    first_last = re.findall(r"([a-zA-Z]*)_(?:[a-zA-Z]*_)?([a-zA-Z]*)_Calib", sdir)
    if first_last: return first_last[0][0] + ' ' + first_last[0][1]
    else: return []
```

```python
def extract_exercise(csv_file):
    # could also extract the first item, which should be the person
#     pers_exer = re.findall(r"([a-zA-Z ]+)(?:_| _)([a-zA-Z]+).{0,7}\.csv", fn)
    exer = re.findall(r"[a-zA-Z ]+(?:_| _)([a-zA-Z]+).{0,7}\.csv", csv_file)
    if exer: return exer[0]
    else: return []
```

```python
def get_best_match(item, possible, cutoff):
    match = difflib.get_close_matches(item, possible, 1, cutoff)
    if match: return match[0] # upack
    else: return []
```

## Build a dictionary, `person:[folder_list]`, labeling the main data files by person
Individual files also contain person names, however the order (first,last or last,first) is inconsistent. So although a fuzzy string search could be used to match individual files to person labels, it's probably more robust to match based on the file names.  
i.e. the names from the main data folders are taken as ground truth (rather than the person names in the CSV files)
>**Warning:**  Renamed one of the data directories since it appears to be the only folder with non-matching person names.
`Rick_`**`R`**`_Calibration_Trial_Fitness5_Push4/` -> `Rick_`**`Rondeau`**`_Calibration_Trial_Fitness5_Push4` 

```python
old = os.path.join(DATA_DIR, 'Rick_R_Calibration_Trial_Fitness5_Push4')
new = os.path.join(DATA_DIR, 'Rick_Rondeau_Calibration_Trial_Fitness5_Push4')
try:
    os.rename(old, new)
except:
    print old, 'does not exist or was already renamed'
```

```python
sub_dirs = os.listdir(DATA_DIR)

personList = []
personSet = set() # sets are automatically unique
for sdir in sub_dirs:
    # pattern is: <first>_<middle(optional)>_<last>_Calibration...
    name = extract_name(sdir)
    personList.append(name)        # may be empty
    if name: personSet.add(name)
    
print len(personSet), 'unique persons:\n', personSet
```

Traverse again to create a dictionary of top level data directories based on person keys.  
If the directories can't be parsed according to `get_best_match()`, ignore them.


```python
ignoredFolders = [];
personDict = {}
for per in personSet:
    personDict[per] = []

cutoff = 0.9
for i, pl in enumerate(personList): # same length as sub_dirs
    if pl:
        match = get_best_match(pl, personSet, cutoff)
        personDict[match].append(sub_dirs[i])
    else:
        ignoredFolders.append(sub_dirs[i])
        print 'ignored folder: ', sub_dirs[i]
        
# write to file
with open(os.path.join(CLEAN_DIR,'person_dirs.txt'), 'w') as outfile:
    json.dump(personDict, outfile,
              indent=2)
outfile.close()
```

## Read unique exercises from manually created list
Too many unique exercises exist due to variety and poor naming conventions.  
We'll have to depend on being able to match the exercise names in the files with a fuzzy search

```python
exercise_set = set()
f = open(EXER_LIST);
for line in f:
    exercise_set.add(line.rstrip())
f.close()
```

## Create dictionary of dictionaries, `persons: exercises: [files]`
Build dictionaries from the existing person and exercise sets.
Walk through the tree to find all .csv files.
For each file, parse the person and exercise names, and match to dictionary keys.

```python
personExerDict = {}
for per in personSet:
    personExerDict[per] = {ex:[] for ex in exercise_set}
```

```python
cutoff = 0.5
count = 0
count_ign = 0
ignoredFiles = []
for p in personSet:
    for sdir in personDict[p]:   # valid subdirectories of the top-level directory
        spath = os.path.join(DATA_DIR, sdir)
        for root, dirnames, fnames in os.walk(spath):   # all files under eash subdirectory
            for fn in fnmatch.filter(fnames, '*.csv'):
                exer = extract_exercise(fn)
                csvPath = os.path.join(root, fn)
                if exer:
                    e = get_best_match(exer, exercise_set, cutoff)
                    if e:
                        personExerDict[p][e].append(csvPath)
                        count += 1
                    else:
                        ignoredFiles.append(csvPath)
#                         print 'ignored file: ', csvPath
                        count_ign += 1
                else:
                    ignoredFiles.append(csvPath)
#                     print 'ignored file: ', csvPath
                    count_ign += 1

with open(os.path.join(CLEAN_DIR,'person_exer_dirs.txt'), 'w') as outfile:
    json.dump(personExerDict, outfile,
              indent=2)
outfile.close()

print count_ign, 'of', count, 'files ignored'
```

## Count my exercises

```python
counts = {e:0 for e in exercise_set}
# count for each person
for p in personSet:
    for e in exercise_set:
        counts[e] += len(personExerDict[p][e])
# print counts
```

Order the dictionary (using OrderedDict) according to value, from the greatest down.

```python
from collections import OrderedDict
counts_ord = OrderedDict(sorted(counts.items(), key=lambda (_,value): value, reverse=True))
```
